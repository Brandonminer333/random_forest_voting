# Random Forest Classifier Voting

```python
from ClassifierVoting import *


X_train, X_test, Y_train, Y_test = make_random_dataset(n=1000, p=6, strength=0.5, noise=0.5, bins=6, random_state=42)

rf = RandomForestClassifier(n_estimators=100, n_jobs=-1)

rf.fit(X_train, Y_train)

train_preds = get_predictions(rf, X_train)
test_preds = get_predictions(rf, X_test)
```

```python
type(train_preds)
```

```python
class_proportions(Y_train).sort_values(by='proportion')
```

## Majority Vote

This is the easiest to understand method. Find the mode of every tree, then find the mode of the tree modes.

```python
acc, confusion = accuracy(train_preds, Y_train, 'majority')
print(acc)
```

```python
acc, confusion = accuracy(test_preds, Y_test, 'majority')
print(acc)
display(confusion)
```

## Weighted Majority / Probability

```python
acc, confusion = accuracy(train_preds, Y_train, 'weighted')
print(acc)
```

```python
acc, confusion = accuracy(test_preds, Y_test, 'weighted')
print(acc)
display(confusion)
```

## Ranked Voting

```python
acc, confusion = accuracy(train_preds, Y_train, 'ranked')
print(acc)
```

```python
acc, confusion = accuracy(test_preds, Y_test, 'ranked')
print(acc)
display(confusion)
```

## Takeaway

The Cumulative approach performs the best given a fine-tuned random forest, while the majority/ranked approach does better given a high `min_leaf_sample`.
